```python
import os
import numpy as np
import pandas as pd
import mne
import matplotlib.pyplot as plt
from autoreject import AutoReject
from mne.preprocessing import ICA, create_eog_epochs, create_ecg_epochs
%matplotlib qt
```

```python
montage = mne.channels.make_standard_montage("GSN-HydroCel-128")
montage.plot()
```

```python
raw = mne.io.read_raw_egi('C:/Learn/Project/bylw/eeg/lal-hc-405-task_20230825_084820.mff', preload=True)
raw_copy = raw.copy()
```

```python
# 查看所有事件类型及其对应的数量
events = mne.events_from_annotations(raw)

# 目标 event_id 对应的数值（确保 '2   ' 这个键在 event_id 字典中）
target_event_name = '2   '  # 确保这里的字符串格式正确
if target_event_name in events[1]:
    target_event_id = events[1][target_event_name]
    
    # 统计该 event_id 在 events[:, 2]（第三列）中出现的次数
    count = np.sum(events[0][:, 2] == target_event_id)
    
    print(f"Event '{target_event_name}' (ID {target_event_id}) 出现了 {count} 次")
else:
    print(f"事件 '{target_event_name}' 不在 event_id 字典中，请检查名称是否正确")
```

```python
# 去除channel
raw_copy.drop_channels(['E1', 'E2', 'E8','E14', 'E15', 'E17', 'E21', 'E25', 'E26', 'E32', 'E38', 'E39', 'E43', 'E44', 'E45', 'E48', 'E49', 'E108', 'E113', 'E114', 'E115', 'E119', 'E120', 'E121', 'E125', 'E126', 'E127', 'E128'])
# raw_copy.ch_names
eogs = {item:'eog' for item in [ 'E9', 'E22',]}
raw_copy.set_channel_types({'VREF': 'misc'})
raw_copy.set_channel_types(eogs)
```

```python
raw
```

```python
raw_copy.set_montage(montage)
fig = raw_copy.plot_sensors(show_names=True)
```

```python
# 重参考
raw_copy.set_eeg_reference('average', ch_type='eeg')
raw_copy.info['custom_ref_applied']
```

```python
# 滤波
picks = mne.pick_types(raw_copy.info, eeg=True, eog=False)
raw_copy.filter(l_freq=0.1, h_freq=30, fir_design='firwin', phase='zero-double', picks=picks)
```

```python
raw_copy.resample(100)
```

```python
# 截取
epochs = mne.Epochs(raw_copy, events[0], event_id=17, tmin=-0.2, tmax=0.5, baseline=None, preload=True)
epochs
```

```python
# ica
ica = mne.preprocessing.ICA(n_components=0.99, random_state=97, method="infomax", max_iter=3000)
ica.fit(raw_copy)

eog_inds, scores = ica.find_bads_eog(epochs)
print(eog_inds)

epochs_clean = ica.apply(epochs.copy())
```

```python
epochs_clean.average().plot_joint()
```

```python
# 自动删除坏导
# 初始化AutoReject，可自定义参数（如n_interpolate控制最大插值通道数）
ar = AutoReject()
epochs_clean = ar.fit_transform(epochs)

# 查看被拒绝的epoch和插值情况
reject = get_rejection_threshold(epochs)
reject
```
